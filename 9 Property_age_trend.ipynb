### How does YearBuilt (age of property) affect price? Is there a "sweet spot" age? 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:\\Users\\pc\\OneDrive\\Desktop\\Melbourne_Housing_data_analyst\\melbourne_housing_clean.csv')

df['Age'] = df['YearBuilt'].dropna().apply(lambda x: 2017 - x) # expacting data is from 2017

df['Age']
```

```python

# Set style for better aesthetics
sns.set_style("whitegrid")
plt.figure(figsize=(14, 8))

# Create the lineplot with enhancements
lineplot = sns.lineplot(
    data=df[df['Age'] < 200],  # Filter to only include properties with age less than 200
    x='Age',
    y='Price',
    estimator='mean',  # Show mean price at each Age
    errorbar=('ci', 95),  # 95% confidence interval
    linewidth=3,
    color='royalblue',
    markersize=8,
    markeredgecolor='white',
    markeredgewidth=1.5
)

# Add title and labels with better formatting
plt.title('Property Price Trend by Age', fontsize=20, fontweight='bold', pad=20)
plt.xlabel('Age of Property (Years)', fontsize=14, fontweight='semibold')
plt.ylabel('Price ($)', fontsize=14, fontweight='semibold')

# Format y-axis as currency
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

# Add grid for better readability
plt.grid(True, alpha=0.3, linestyle='--')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
